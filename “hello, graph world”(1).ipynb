```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_edges_from([(1,2),(2,1),(3,1),(4,6),(5,6),(5,7),
                  (1,3),(2,5),(3,2),(4,7),(6,7),(7,8),
                  (1,4),(2,1),(3,5),(4,5),(8,9),(9,10),
                  (1,5),(3,8),(3,7),(4,1)
                  ])

cities = {0:"Carol",1:"Andre",2:"Fernando",3:"Diane"
         ,4:"Beverly",5:"Ed",6:"Garth",7:"Heather"
         ,8:"Ike",9:"Jane"}
H=nx.relabel_nodes(G,cities)
nx.draw(H, with_labels = True)
```
